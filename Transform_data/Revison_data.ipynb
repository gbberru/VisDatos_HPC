```python
import pandas as pd
import pyarrow.parquet as pq
from pathlib import Path

# Ruta de tu dataset final reducido
DATASET_PATH = Path(r"D:\GB TI\Master\Visu_DATOS\Proyecto\project_hpc_visdatos\VisDatos_HPC\data\raw\datacenter_enriched_core_single.parquet")

# Leer metadata sin cargar todo el dataset
pf = pq.ParquetFile(DATASET_PATH)

print("Archivo:")
print(DATASET_PATH)

print("\nNúmero de filas:")
print(pf.metadata.num_rows)

print("\nNúmero de columnas:")
print(len(pf.schema.names))

print("\nColumnas del dataset:")
for i, col in enumerate(pf.schema.names, start=1):
    print(f"{i}. {col}")
```

```python
# Cargar dataset
df_core = pd.read_parquet(DATASET_PATH)

print("Dimensiones del dataset:")
print(df_core.shape)

print("\nPrimeras filas:")
display(df_core.head())

print("\nÚltimas filas:")
display(df_core.tail())
```

```python
# Revisa tipos de datos
print("Tipos de datos por columna:")
display(df_core.dtypes.reset_index().rename(columns={"index": "columna", 0: "tipo_dato"}))
```

```python
# Revisa valores nulos
null_summary = pd.DataFrame({
    "columna": df_core.columns,
    "nulos": df_core.isna().sum().values,
    "porcentaje_nulos": (df_core.isna().mean().values * 100).round(2)
})

null_summary = null_summary.sort_values("porcentaje_nulos", ascending=False)

print("Resumen de valores nulos:")
display(null_summary)
```

```python
# Revisa columnas numéricas principales
numeric_cols = df_core.select_dtypes(include=["number"]).columns

print("Columnas numéricas:")
print(list(numeric_cols))

print("\nEstadísticas descriptivas:")
display(df_core[numeric_cols].describe().T)
```

```python
# Revisa columnas categóricas 
categorical_cols = df_core.select_dtypes(include=["object", "string", "category"]).columns

print("Columnas categóricas:")
print(list(categorical_cols))

for col in categorical_cols:
    print(f"\nColumna: {col}")
    print(df_core[col].value_counts(dropna=False).head(20))
```

```python
# Revisa el rango temporal
df_core["timestamp"] = pd.to_datetime(df_core["timestamp"], errors="coerce")
df_core["start_date"] = pd.to_datetime(df_core["start_date"], errors="coerce")
df_core["end_date"] = pd.to_datetime(df_core["end_date"], errors="coerce")

print("Rango temporal de mediciones:")
print(df_core["timestamp"].min(), "→", df_core["timestamp"].max())

print("\nRango de inicio de jobs:")
print(df_core["start_date"].min(), "→", df_core["start_date"].max())

print("\nRango de fin de jobs:")

print(df_core["end_date"].min(), "→", df_core["end_date"].max())
```

```python
# Validación de timestamp dentro del job
df_core["timestamp_dentro_del_job"] = (
    (df_core["timestamp"] >= df_core["start_date"]) &
    (df_core["timestamp"] <= df_core["end_date"])
)

print("Validación timestamp dentro del job:")
print(df_core["timestamp_dentro_del_job"].value_counts(dropna=False))
```

```python
print("Cantidad de nodos únicos:")
print(df_core["node"].nunique())

print("\nPrimeros nodos:")
print(df_core["node"].drop_duplicates().head(30).to_list())

print("\nRacks inferidos:")
print(df_core["rack_inferred"].value_counts(dropna=False).head(20))

print("\nDistribución CPU/GPU:")
print(df_core["gpu_node"].value_counts(dropna=False))

print("\nModelos GPU:")
print(df_core["gpu_model"].value_counts(dropna=False).head(20))

print("\nCantidad de GPU por nodo:")
print(df_core["gpu_count"].value_counts(dropna=False))
```

```python

print("Cantidad de jobs SLURM únicos:")
print(df_core["slurm_id"].nunique())

print("\nEstados de jobs:")
print(df_core["state"].value_counts(dropna=False))

print("\nTipos de nodo usados en jobs:")
print(df_core["nodetypes"].value_counts(dropna=False).head(20))

print("\nNúmero de nodos por job:")
display(df_core["numnodes"].describe())

print("\nNúmero de cores por job:")
display(df_core["numcores"].describe())

print("\nDuración de jobs en horas:")
display(df_core["job_duration_hours"].describe())
```

```python
# Revisa columnas numéricas principales
cols_to_check = [
    "node_power_usage",
    "node_rapl_package_power_sum",
    "node_energy_kwh_est",
    "cpu_package_energy_kwh",
    "co2_kg_est",
    "node_load1",
    "node_load5",
    "node_load15",
    "node_load1_per_core",
    "ram_active_gb",
    "ram_free_gb",
    "node_hwmon_temp_celsius_mean",
    "node_hwmon_temp_celsius_max",
    "node_thermal_zone_temp_mean",
    "node_thermal_zone_temp_max",
    "gpu_power_watts_mean",
    "gpu_power_watts_max",
    "gpu_memory_used_gb",
    "nvidia_gpu_temperature_celsius_mean",
    "nvidia_gpu_temperature_celsius_max",
    "nvidia_gpu_duty_cycle_mean",
    "nvidia_gpu_duty_cycle_max"
]

existing_cols_to_check = [col for col in cols_to_check if col in df_core.columns]

summary_check = df_core[existing_cols_to_check].describe().T

display(summary_check)
```

```python
# Revisa valores negativos en columnas numéricas principales
non_negative_cols = [
    "node_power_usage",
    "node_energy_kwh_est",
    "cpu_package_energy_kwh",
    "co2_kg_est",
    "ram_active_gb",
    "ram_free_gb",
    "gpu_power_watts_mean",
    "gpu_power_watts_max",
    "gpu_memory_used_gb",
    "node_disk_read_bytes_total_sum_delta",
    "node_disk_written_bytes_total_sum_delta",
    "node_network_receive_bytes_total_sum_delta",
    "node_network_transmit_bytes_total_sum_delta"
]

non_negative_cols = [col for col in non_negative_cols if col in df_core.columns]

print("Valores negativos por columna:")
for col in non_negative_cols:
    negatives = (df_core[col] < 0).sum()
    print(f"{col}: {negatives}")
```

```python
# Revisa duplicados
print("Filas duplicadas completas:")
print(df_core.duplicated().sum())

print("\nDuplicados por prom_id:")
if "prom_id" in df_core.columns:
    print(df_core["prom_id"].duplicated().sum())
```
